```python
import matplotlib.pyplot as plt
import numpy as np
```

# Plots

```python
y = np.array([0, 10])

plt.plot(y)           # The default values for x-axis are 0, 1, 2, 3 ....
plt.show()
```

```python
plt.plot(y, linewidth = "10")           # The default values for x-axis are 0, 1, 2, 3 ....
plt.show()
```

```python
x = np.array([0, 20])
plt.plot(x, y)
plt.show()
```

```python
x = np.array([0, 2, 4, 6, 8])
y = np.array([0, 4, 16, 36, 64])

plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, marker = 'o')
plt.show()
```

```python
plt.plot(x, y, marker = "x")
plt.show()
```

```python
plt.plot(x, y, marker = "s", ms = 10, mec = "g", mfc = "r")         # "ms" is marker size, "mec" is marker edge color, "mfc" is marker face color
plt.show()
```

```python
plt.plot(x, y, "o--g")          # marker is "o" and "--" is for dashed line and "g" is for green color
plt.show()
```

```python
y1 = np.array([1, 3, 5, 7])
y2 = np.array([0, 4, 8, 12])

plt.plot(y1)
plt.plot(y2)

plt.show()        # Two plots on same graph
```

```python
x = np.array([0, 2, 4, 6, 8])
y = np.array([0, 4, 16, 36, 64])

plt.title("Matplotlib Pyplot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.plot(x, y)
plt.show()          # with title, xlabel and ylabel
```

```python
plt.title("Matplotlib Pyplot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.plot(x, y)
plt.grid()
plt.show() 
```
